**Import the relevant libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
from collections import Counter
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
from xgboost import plot_importance
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

warnings.filterwarnings('ignore')
```

**Load the data**

```python
raw_data = pd.read_excel(r'C:\Users\user\Downloads\a_Dataset_CreditScoring.xlsx')
```

```python
raw_data.head()
```

**Data preparation**

```python
#show count of rows and columns
raw_data.shape
```

```python
#make a copy of the raw data
data = raw_data.copy()
data
```

```python
#drop the customer ID column
data = data.drop('ID', axis=1)
data
```

```python
#explore missing values
data.isnull().sum()
```

```python
#fill missing values with mean
data = data.fillna(data.mean())
```

```python
#explore missing values after fix
data.isnull().sum()
```

```python
#count of good loans(0) and bad loans(1)
#the count shows there is class imbalance
data['TARGET'].value_counts()
```

```python
#data summary across 0 & 1
#here i grouped the data by the target and checked the mean of each feature
#this helps to spot trends e.g do bad loan customers have lower income? or do they have more debt?
#this is useful for early feature selection and understanding model behaviour
data.groupby('TARGET').mean()
```

```python
data.head()
```

**Declare the dependent and independent variable**

```python
y = data['TARGET']
x = data.drop('TARGET', axis=1)
```

```python
#scale the data
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
#get the standardized inputs
x_scaled = scaler.transform(x)
```

**Split data into test and training**

```python
x_scaled_train, x_scaled_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, stratify=y, random_state=42)
```

**Handle Class Imbalance**

**Create a SMOTE object**

```python
smote = SMOTE(random_state=42)
```

```python
# Apply SMOTE to the training features and labels
x_scaled_train_sm, y_train_sm = smote.fit_resample(x_scaled_train, y_train)
```

**Check the new class balance**

```python
print("Before SMOTE:", Counter(y_train))
print("After SMOTE:", Counter(y_train_sm))
```

**XGBoost**

```python
#initialize xgboost with imbalance handling
xgb_model = XGBClassifier(
    scale_pos_weight=2,  # Helps with imbalance
    use_label_encoder=False, 
    eval_metric='logloss',
    random_state=42
)
```

```python
#fit the model
xgb_model.fit(x_scaled_train_sm, y_train_sm)
```

**Make prediction on the test set**

```python
y_pred_xgb = xgb_model.predict(x_scaled_test)
```

**Threshold tuning**

```python
#get predicted probabilities for class 0 and 1
y_probs = xgb_model.predict_proba(x_scaled_test)
```

```python
# Save both columns in a DataFrame
df_prob = pd.DataFrame(y_probs, columns=["Prob_Class_0", "Prob_Class_1"])
```

```python
#set a lower threshold
y_pred_thresh = (df_prob["Prob_Class_1"] > 0.25).astype(int)
```

```python
y_pred_thresh.head()
```

**Model performance**

```python
#confusion matrix
cm = confusion_matrix(y_test, y_pred_thresh)
print("Confusion Matrix:")
print(cm)
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Pred 0', 'Pred 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
#accuracy score
print(accuracy_score(y_test, y_pred_thresh))
```

**Classification report**

```python
# Generate the report as a dictionary
report_dict = classification_report(y_test, y_pred_thresh, output_dict=True, target_names=['good loan', 'bad loan'])

# Convert to a DataFrame
report_df = pd.DataFrame(report_dict).transpose()

# Round for cleaner display
report_df = report_df.round(3)

# Show the table
report_df
```

**Writing output file**

```python
#Create DataFrames
df_actual = pd.DataFrame(y_test.values, columns=["Actual"])
df_prob = pd.DataFrame(y_probs, columns=["Prob_Class_0", "Prob_Class_1"])
df_pred = pd.DataFrame(y_pred_thresh.values, columns=["Predicted"])
```

```python
# Concatenate into final result
df_output = pd.concat([df_actual.reset_index(drop=True), df_prob, df_pred], axis=1)
```
